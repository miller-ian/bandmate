```python
from midiutil import MIDIFile

# Set up the MIDI file parameters
tempo = 180
time_signature = (4, 4)
clocks_per_tick = 24
numberOf32ndNotesPer24Clocks = 8
track = 0
channel = 0
volume = 100
duration = 1

#create a new MIDIFile object with one track
midi_file = MIDIFile(numTracks=1)

#add a track to the MIDIFile object
midi_file.addTrackName(track, 0, "Ian Melody")
midi_file.addTempo(track, 0, tempo)
midi_file.addTimeSignature(track, 0, *time_signature, clocks_per_tick, numberOf32ndNotesPer24Clocks)


# prompt to GPT:
melody = [
    (40, 0),   # C4 at start
    (30, 1),   # Eb4 after a beat
    (37, 2),   # F4 after two beats
    (33, 3),   # D4 after three beats
    
]
#add each note to the MIDI file
for note, start_time in melody:
    midi_file.addNote(
        track, channel, note, start_time, duration, volume
    )

#write the MIDI file to disk
with open("my_melody.mid", "wb") as output_file:
    midi_file.writeFile(output_file)
```

```python
from midi2audio import FluidSynth

#convert the MIDI file to an audio file
FluidSynth(sound_font="gm.sf2").midi_to_audio("my_melody.mid", "my_melody.wav")
```
